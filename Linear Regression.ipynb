```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("C2 - Sheet1 (1).csv")
X=data.iloc[:,1]
Y=data.iloc[:,2]
plt.scatter(X,Y)
plt.show()

class LinearRegression:
    def __init__(self):
        self.learning_rate=0.0001
        self.epochs=1500
        self.gradient=0
        self.bias=0
        self.n=(len(Y))

    
    def MSE(self):
        Error=0
        for i in range(self.n):
            Error+=(1/2*self.n)*(Y[i]-self.gradient*X[i]+self.bias)**2
        print(f"Cost:{Error/self.n}")
    
    def gradient_descent(self):
        m=0
        b=0
        for i in range(self.n):
            m+=(-2/self.n)*X[i]*(Y[i]-self.gradient*X[i]+self.bias)
            b+=(-2/self.n)*(Y[i]-self.gradient*X[i]+self.bias)
            
        self.gradient=self.gradient-self.learning_rate*m
        self.bias=self.bias-self.learning_rate*b
        
    
    def iterator(self):
        for i in range(self.epochs):
            self.MSE()
            self.gradient_descent()
            Eq=self.gradient*X+self.bias
            print(f"Bias:{self.bias},Gradient:{self.gradient}")
        plt.scatter(X,Y)
        plt.scatter(X,Eq)
        plt.show()
           
            


lin=LinearRegression()
lin.iterator()
            
                    
        




    
    
        
    
        
        
        
    
```
